***Data Cleaning Process***

```python
# let's perform the data cleaning task
# Primarily We Will Check the null Values in each colunns
Fedex_df.isnull().sum()              # As we can see here Columns "Shipment Mode", "Dosage" and "Line Item Insurance (USD)" contains the null values
```

```python
for shipment_mode, count in Fedex_df['Shipment Mode'].value_counts().items():
  print(f"{shipment_mode} : {count}")
```

```python
# As we can see here large no of Shipment made by "Air", So Here we will replace all the null values with "Air" in Shipment Mode Column
Fedex_df['Shipment Mode'] = Fedex_df['Shipment Mode'].fillna('Air')
```

```python
# Similarly Do Here for "Dosage" Column
Mode_of_Dosage = Fedex_df['Dosage'].mode()[0]
Fedex_df['Dosage'].fillna(Mode_of_Dosage, inplace = True)
```

```python
Fedex_df['Line Item Insurance (USD)'].unique()
```

```python
# For "Line Item Insurance (USD)" Column first we will find the mean of it and then replace all nan values with it
Fedex_df['Line Item Insurance (USD)'].fillna(Fedex_df['Line Item Insurance (USD)'].mean(), inplace=True)
```
